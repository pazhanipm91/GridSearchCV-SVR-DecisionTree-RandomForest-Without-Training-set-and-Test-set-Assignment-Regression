```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the dataset
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
indep
```

```python
dep
```

```python
#Removed the Training Set and the Test Set here
```

```python
from sklearn.ensemble import RandomForestRegressor
```

#Removed the Standard Scaler options as well

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['mse','friedman_mse'],
              'max_features':['auto', 'sqrt', 'log2'],
              'n_estimators':[10,100]} #Here We used the Dictionary methods = {}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3, n_jobs=-1) # Up to this code is just a procedure

#fitting the model for grid search CV
grid.fit(indep, dep)
```

```python
# Here we used this GRID SEARCH method SVR(), So like wise we have to check with other models as well.
```

```python
# Print the best parameter after tuning 
# Print (grid.best_params_)
re=grid.cv_results_

print("The R_score value for the best parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
